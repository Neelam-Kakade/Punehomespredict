```python
def convert(x):
    pp = x.split('-')
#     print("pp--->",tokel)
    if len(pp) == 2:
        return(float(pp[0])+float(pp[1]))/2
    try:
        return float(x)
    except:
        return None
df['new_total_sqft']=df['total_sqft'].apply(convert)
df['new_total_sqft']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x=df.drop("price",axis=1)
x
```

```python
y=df["price"]
y
```

```python
vif_list = []

for i in range(x.shape[1]):
    vif = variance_inflation_factor(x.to_numpy(),i)
    vif_list.append(vif)
    
s1 = pd.Series(vif_list, index=x.columns)
s1.sort_values().plot(kind = 'barh') # sort function to get graph properly in order (bar for vertical
```

```python
# Model Training
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
x_train
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred_test = model.predict(x_test)
y_pred_test
print(len(y_pred_test))
```

```python
y_pred_train = model.predict(x_train)
print(y_pred_train)
print(len(y_pred_train))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Residual on train for Normality
residual_train = y_train - y_pred_train # yactual - ypredicated
residual_train
```

```python
# Residual on test for Normality
residual_test = y_test - y_pred_test # yactual - ypredicated
residual_test
```

```python
sns.kdeplot(residual_train,fill=True)
```

```python
sns.kdeplot(residual_test,fill=True)
```

```python
sns.scatterplot(x= y_pred_train, y= residual_train)
```

```python
sns.scatterplot(x= y_pred_test, y= residual_test)
```

```python
# Training Data Set
y_pred_train = model.predict(x_train)

mse = mean_squared_error(y_train, y_pred_train)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("MAE :",mae)

r2 = r2_score(y_train, y_pred_train)
print("R Squared :",r2)
```

```python
# Testing Dataset
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred_test)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_test, y_pred_test)
print("MAE :",mae)

r2 = r2_score(y_test, y_pred_test)
print("R Squared :",r2)
```
